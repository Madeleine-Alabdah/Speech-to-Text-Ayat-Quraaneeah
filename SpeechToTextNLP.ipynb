```python
AYATS_FOLDERS = "/content/drive/MyDrive/SPEECH_TO_TEXT/Ayat-Quraaneah/"
```

```python
def convert_audio(filename):
# files
  filenamesplit = os.path.splitext(filename)[0]
  filenamewav = filenamesplit + '.wav'
  print(filenamewav)
  newfilename = "mono_" + filenamewav
# convert mp3 to wav                                                            
  # sound = AudioSegment.from_m4a(dirname + filename)
  # sound.export(dirname + filenamewav, format="wav")
# In this section we can convert any type such as mp3, m4a to wav to be able to handle the record
  sound = AudioSegment.from_file(AYATS_FOLDERS + filename)
  sound.export(AYATS_FOLDERS + filenamewav, format="wav")
```

We call the function that we created "convert_audio" 

```python
for filename in os.listdir(AYATS_FOLDERS):
  print(filename)
  print("Starting...")
  convert_audio(filename)
  print("Completed!!")
```
